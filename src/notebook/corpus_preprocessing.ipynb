```python
cl = []
l = sorted(os.listdir(f'{base}/data/modnucc/nucc'))
for file in l[:]:
    cl.append(nucc.Corpus(f'{base}/data/modnucc/nucc/{file}'))
```

```python
include_period = 0
all = 0
speakers = set()
for c in cl:
    for s in c.conversation:
        all += 1
        speakers.add(s.speaker)
        if('。' in s.text[:-1]):
            include_period += 1
print(f'全体行数:{all}, 二文以上:{include_period}')
[print(f':{i}') for i in sorted(list(speakers))]
```

```python
import re
new_corpus_text_list = corpus_text_list.copy()
for index, corpus_text in enumerate(new_corpus_text_list):
    corpus_text = re.sub('\n+', '\n', corpus_text)
    corpus_text = corpus_text.replace('＠','@')
    corpus_text = corpus_text.replace('：',': ')
    corpus_text = corpus_text.replace('＜笑い。','＜笑い＞')
    corpus_text = re.sub('＜[^＞]+＞', '', corpus_text)
    corpus_text = re.sub('@.+$', '@END', corpus_text)
    corpus_text = re.sub('（[^）]+）', '', corpus_text)
    corpus_text = re.sub('\n+', '\n', corpus_text)
    new_corpus_text_list[index] = corpus_text
    with open(f'{base}/data/modnucc/newnucc/{corpus_file_list[index]}', mode='w') as f:
        f.write(corpus_text)
    
```
